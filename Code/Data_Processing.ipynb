```python
def calculate_p_one(df, lambda_val, mu_val, c):
    # Calculate traffic intensity rho
    rho = df[lambda_val] / (df[mu_val] * c)
    
    # Calculate probability of zero customers in queue
    p_zero = calculate_p_zero(df, lambda_val, mu_val, c)
    
    # Calculate probability of one customer in queue
    p_one = (rho**c) / np.math.factorial(c) * p_zero
    
    return p_one



# Calculate probabilities for morning with 1 server and 2 servers
df['P1_morning_1_server'] = calculate_p_one(df, 'Arrival Rate Morning', 'Service Rate', 1)
df['P1_morning_2_servers'] = calculate_p_one(df, 'Arrival Rate Morning', 'Service Rate', 2)

# Calculate probabilities for evening with 1 server and 2 servers
df['P1_evening_1_server'] = calculate_p_one(df, 'Arrival Rate Evening', 'Service Rate', 1)
df['P1_evening_2_servers'] = calculate_p_one(df, 'Arrival Rate Evening', 'Service Rate', 2)

# Print the DataFrame to verify the new columns
df
```

```python
class MM1:

    def __init__(self, arrival_rate, service_rate):
        self.arrival_rate = float(arrival_rate)
        self.service_rate = float(service_rate)
        self.rho = self.arrival_rate / self.service_rate
        if self.rho >= 1:
            raise ValueError("System is overloaded. Service rate must be greater than arrival rate.")

    @property
    def capital_p0(self):
        return 1 - self.rho

    def capital_p(self, n):
        return self.rho ** n * self.capital_p0

    def capital_l(self):
        return self.arrival_rate * self.capital_w()

    def capital_lq(self):
        return self.arrival_rate * self.capital_wq()

    def capital_w(self):
        EPS = 1e-10  # Small value for stability
        return 1 / (self.service_rate - self.arrival_rate + EPS)

    def capital_wq(self):
        return self.rho * self.capital_w()

    def set_arrival_rate(self, arrival_rate):
        self.arrival_rate = float(arrival_rate)
        self.rho = self.arrival_rate / self.service_rate

    def set_service_rate(self, service_rate):
        self.service_rate = float(service_rate)
        self.rho = self.arrival_rate / self.service_rate

    def get_arrival_rate(self):
        return self.arrival_rate

    def get_service_rate(self):
        return self.service_rate

    def get_rho(self):
        return self.rho


mm1 = MM1(0.005, 0.05)
```

```python
class MMc:

    def __init__(self, _lambda, mu, c):
        # lambda : mean arrival rate
        self.__lambda = eval(_lambda)

        # mu : mean service rate
        self.__mu = eval(mu)

        # c : means c identical servers
        self.__c = eval(c)

        self.__r = self.__lambda / self.__mu

        self.__rho = self.__r / self.__c

        if self.rho >= 1:
            raise ValueError("System is overloaded. Service rate must be greater than arrival rate.")

    def capital_p0(self):
        acc = 0
        if self.__rho >= 1:
            for i in range(self.__c):
                acc += self.__r ** i / factorial(i)
            acc += (self.__r ** self.__c / factorial(self.__c)) * (self.__c * self.__mu / (self.__c * self.__mu - self.__lambda))
        else:
            for i in range(self.__c):
                acc += self.__r ** i / factorial(i)
            acc += self.__c * self.__r ** self.__c / (factorial(self.__c) * (self.__c - self.__r))
        return 1 / acc

    def capital_p(self, n):
        if 0 <= n < self.__c:
            den = factorial(n)
        else:
            den = self.__c ** (n - self.__c) * factorial(self.__c)
        return self.__r ** n * self.capital_p0() / den

    # L : Expected number of the customers in the system
    def capital_l(self):
        return self.capital_lq() + self.__r

    # Lq : Expected number of the customers in the queue
    def capital_lq(self):
        num = self.__r ** self.__c * self.__mu * self.__lambda
        den = factorial(self.__c - 1) * (self.__c * self.__mu - self.__lambda) ** 2
        return num * self.capital_p0() / den

    # W : Expected waiting time in the system
    def capital_w(self):
        return self.capital_wq() + 1 / self.__mu

    # Wq : Expected waiting time in the queue
    def capital_wq(self):
        return self.capital_lq() / self.__lambda

    # Average number of idle server C_i = c âˆ’ r
    def avg_idle(self):
        return self.__c - self.__r

    def set_lambda(self, _lambda):
        self.__lambda = eval(_lambda)
        self.__r = self.__lambda / self.__mu
        self.__rho = self.__r / self.__c

    def set_mu(self, mu):
        self.__mu = eval(mu)
        self.__r = self.__lambda / self.__mu
        self.__rho = self.__r / self.__c

    def set_c(self, c):
        self.__c = int(c)
        self.__rho = self.__r / self.__c

    def get_mu(self):
        return self.__mu

    def get_lambda(self):
        return self.__lambda

    def get_c(self):
        return self.__c

    def get_rho(self):
        return self.__rho

    def get_r(self):
        return self.__r
```
