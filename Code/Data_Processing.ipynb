# Arrival rate calculation

```python

# Calculating total duration (in hours)
total_duration_morning = 4*60  # Morning shift duration (9am to 1pm)
total_duration_evening = 2*60  # Evening shift duration (4pm to 6pm)

# Calculating arrival rate for each OPD
df['Arrival Rate Morning'] = df['Morning Total'] / total_duration_morning
df['Arrival Rate Evening'] = df['Evening Total'] / total_duration_evening

# Displaying the calculated arrival rates
df[['OPDs', 'Arrival Rate Morning', 'Arrival Rate Evening']]
```

```python
df[df['OPDs'] == 'Med']['Arrival Rate Morning'].mean()
```

```python
listOfOpd = list(df['OPDs'].unique())
listOfOpd
```

```python
for opd in listOfOpd:
    opd_records = df[df['OPDs'] == opd]  # Filter the DataFrame for the current OPD
    mean_morning_arrival_rate = opd_records['Arrival Rate Morning'].mean()  # Calculate the mean arrival rate
    print('The Morning Arrival Rate for', opd, 'per minute is :', mean_morning_arrival_rate)
    print()
    mean_evening_arrival_rate = opd_records['Arrival Rate Evening'].mean()  # Calculate the mean arrival rate
    print('The Evening Arrival Rate for', opd, 'per minute is :', mean_evening_arrival_rate)
    print()
```

```python
def findArrivalRates(data, listofopd):
    for opd in listofopd:
        opd_records = data[data['OPDs'] == opd]  # Filter the DataFrame for the current OPD
        mean_morning_arrival_rate = opd_records['Arrival Rate Morning'].mean()  # Calculate the mean morning arrival rate
        print('The Morning Arrival Rate for', opd, 'per hour is :', mean_morning_arrival_rate)
        print()
        mean_evening_arrival_rate = opd_records['Arrival Rate Evening'].mean()  # Calculate the mean evening arrival rate
        print('The Evening Arrival Rate for', opd, 'per hour is :', mean_evening_arrival_rate)
        print()

# Example usage:
findArrivalRates(df, listOfOpd)
```

# Service Rate

```python
df['OPDs'].unique()
```

```python
average_service_times = {
    'Med': 4,
    # 'Resp': 4,
    'SURGERY': 8,
    'ORTHO': 14,
    'SKIN': 4,
    'ENT OPD': 8,
    'OPTHALAM': 30,
    'PAED OPD': 10,
    'ObGy': 10,
    # 'Reg': 4,
    # 'Pharma': 4
}
average_service_times
```

```python

# Calculate service rate (patients served per minute) for each OPD
service_rates = {opd: 1 / average_service_times[opd] for opd in average_service_times}
service_rates
```

```python
# Display service rates
print("Service Rates for OPD Consultation:")
for opd, rate in service_rates.items():
    print(f"{opd}: {rate:.2f} patients/min")
```

```python
# Convert service rate from patients per minute to patients per hour
service_rates_per_hour = {opd: rate * 60 for opd, rate in service_rates.items()}

# Display service rates per hour
print("Service Rates for OPD Consultation (per hour):")
for opd, rate in service_rates_per_hour.items():
    print(f"{opd}: {rate:.2f} patients/hour")
```

```python
# Add a new column for service rate in the DataFrame
df['Service Rate'] = df['OPDs'].map(service_rates)

# Display the updated DataFrame with the new column for service rate
print(df[['OPDs', 'Service Rate']])
```

```python
def calculate_p_one(df, lambda_val, mu_val, c):
    # Calculate traffic intensity rho
    rho = df[lambda_val] / (df[mu_val] * c)
    
    # Calculate probability of zero customers in queue
    p_zero = calculate_p_zero(df, lambda_val, mu_val, c)
    
    # Calculate probability of one customer in queue
    p_one = (rho**c) / np.math.factorial(c) * p_zero
    
    return p_one



# Calculate probabilities for morning with 1 server and 2 servers
df['P1_morning_1_server'] = calculate_p_one(df, 'Arrival Rate Morning', 'Service Rate', 1)
df['P1_morning_2_servers'] = calculate_p_one(df, 'Arrival Rate Morning', 'Service Rate', 2)

# Calculate probabilities for evening with 1 server and 2 servers
df['P1_evening_1_server'] = calculate_p_one(df, 'Arrival Rate Evening', 'Service Rate', 1)
df['P1_evening_2_servers'] = calculate_p_one(df, 'Arrival Rate Evening', 'Service Rate', 2)

# Print the DataFrame to verify the new columns
df
```
